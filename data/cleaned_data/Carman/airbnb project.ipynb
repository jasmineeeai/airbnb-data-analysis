```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Nicer default style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.dpi'] = 100

df = pd.read_csv("airbnb_listings.csv")
print('Shape:', df.shape)
df.head()
```

```python
#Date Frame . Inspect structure and missing values

df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
## Standardize column names

df.columns = (
    df.columns
    .str.strip()
    .str.title()
    .str.replace(' ', '_'))

df.columns.tolist()
```

```python
## Normalize Missing Values

missing_values = ['NULL', 'N/A', 'unknown', '-', '', ' '] 
df = df.replace(missing_values, np.nan)

df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df.columns
```

```python
text_cols = [
    col for col in df.columns
    if df[col].dtype == 'object'
]

for col in text_cols:
    df[col] = df[col].astype('string').str.strip()

text_cols[:10]
```

```python
if 'Host_Location' in df.columns:
    Location = {
        'København, Denmark': 'Copenhagen, Denmark',
        'København N, Denmark': 'Copenhagen, Denmark',

    }
    df['Host_Location'] = df['Host_Location'].replace(Location)
    print(df['Host_Location'].value_counts(dropna=False).head(20))
else:
    print('Column country not found.')
```
