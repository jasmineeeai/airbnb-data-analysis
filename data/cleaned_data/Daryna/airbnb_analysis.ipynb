```python
import pandas as pd
```

```python
df = pd.read_csv('listings.csv')
```

```python
df.info()
```

```python
# Drop entirely empty columns
columns_to_drop = ['neighbourhood_group', 'license']
```

```python
# Fix corrupted Danish neighborhood names
copenhagen_districts = {
    'sterbro': 'Østerbro',
    'Nrrebro': 'Nørrebro',
    'Vanlse': 'Vanløse',
    'Brnshj-Husum': 'Brønshøj-Husum',
    'Amager st': 'Amager Øst'
}
df['neighbourhood'] = df['neighbourhood'].replace(copenhagen_districts)
```

```python
# Handle missing values for review columns
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['last_review'] = df['last_review'].fillna('No reviews')
```

```python
# Remove rows with missing essential metadata
df = df.dropna(subset=['name', 'host_name'])
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
# Save the general cleaned dataset (keeps all 22k+ rows for geo/room analysis)
df.to_csv('listings_general_cleaned.csv', index=False)
print(f"General cleaned dataset shape: {df.shape}")

# Create and save a dedicated dataset for price analysis (filters out missing/zero prices)
price_df = df.dropna(subset=['price'])
price_df = price_df[price_df['price'] > 0]
price_df.to_csv('listings_price_cleaned.csv', index=False)
print(f"Price-filtered dataset shape: {price_df.shape}")
```
